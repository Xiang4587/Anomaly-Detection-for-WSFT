### Seperate passing/failing in the dataset

```python
## The class we want to train
Pass_data_x = []
Pass_data_y = []

## Outlier
Fail_data_x = []
Fail_data_y = []

for k in range(len(DF)):
    if DF['hardbin_FT1'][k] == 1:
        feature_tmp = []
        for s in features.keys():
            feature_tmp.append(features[s][k])
        Pass_data_x.append(feature_tmp)
        Pass_data_y.append(DF['hardbin_FT1'][k])
    else:
        feature_tmp = []
        for s in features.keys():
            feature_tmp.append(features[s][k])
        Fail_data_x.append(feature_tmp)
        Fail_data_y.append(DF['hardbin_FT1'][k])

Pass_data_x = np.array(Pass_data_x)
Pass_data_y = np.array(Pass_data_y)
Fail_data_x = np.array(Fail_data_x)
Fail_data_y = np.array(Fail_data_y)
```

### Split train/ test/ abnormal dataset

* Training dataset: train the model 
* Testing dataset: test the model, whether it classifies the passing data
* Abnormal dataset: let model detect the outliers

```python
def splitData(Data_x, Data_y, ratio):
    num = int(round(len(Data_x)*ratio))
    par = len(Data_x) - num
    x_main = Data_x[:num]
    y_main = Data_y[:num]
    x_mini = Data_x[-par:]
    y_mini = Data_y[-par:]
    return x_main, y_main, x_mini, y_mini
```

```python
## Shuffle the dataset
def unison_shuffled_copies(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]
```

```python
Pass_data_x, Pass_data_y = unison_shuffled_copies(Pass_data_x, Pass_data_y)
```

```python
X_train, Y_train, X_test, Y_test = splitData(Pass_data_x, Pass_data_y, 0.85)
X_outlier = Fail_data_x
Y_outlier = Fail_data_y
Y_outlier_re = np.array([-1 for i in range(len(Y_outlier))]) ## Take it as a binary classification problem
```

```python
print("Train:", len(X_train))
print("Test:", len(X_test))
print("Outlier:", len(X_outlier))
```

```python
# fit the model
osvm_clf = svm.OneClassSVM(nu=0.09, kernel='rbf', gamma=0.25)
osvm_clf.fit(X_train)
```

## Evaluate model

```python
Y_outlier_re = np.array([-1 for i in range(len(Y_outlier))])
confusion_outlier = confusion_matrix(Y_outlier_re, y_pred_outlier)
confusion_test = confusion_matrix(Y_test, y_pred_test)
confusion_mat = confusion_outlier + confusion_test
```

```python
confusion_mat
```

```python
## Metrics: precision, recall and F-index
tp = float(confusion_mat[0][0])
fn = float(confusion_mat[0][1])
fp = float(confusion_mat[1][0])
tn = float(confusion_mat[1][1])
pre = tp/(tp+fp)
recall = tp/(tp+fn)
F_index = 2*(pre*recall)/(pre+recall)
print("Precision:", pre)
print("Recall:", recall)
print("F-index:", F_index)
```

## Feature selection

### Recursive feature elimination

```python
def buildModel(X_train, nu=0.09, gamma=0.25):
    osvm_clf = svm.OneClassSVM(nu=nu, kernel='rbf', gamma=gamma)
    osvm_clf.fit(X_train)
    return osvm_clf
```

```python
def ConfusionMatrix(y_test_true, y_test_pre, y_outlier_true, y_outlier_pre):
    from sklearn.metrics import confusion_matrix
    confusion_outlier = confusion_matrix(y_outlier_true, y_outlier_pre)
    confusion_test = confusion_matrix(y_test_true, y_test_pre)
    confusion_mat = confusion_outlier + confusion_test
    return confusion_mat
```

```python
def evaluate(cm):
    tp = float(cm[0][0])
    fn = float(cm[0][1])
    fp = float(cm[1][0])
    tn = float(cm[1][1])
    pre = tp/(tp+fp)
    recall = tp/(tp+fn)
    F_index = 2*(pre*recall)/(pre+recall)
    return [pre, recall, F_index]
```

```python
Rank = {}
for f in range(len(features.keys())):
    X_train_rm = np.delete(X_train, f, 1)
    X_test_rm = np.delete(X_test, f, 1)
    X_outlier_rm = np.delete(X_outlier, f, 1)
    
    clf = buildModel(X_train_rm) ## Model
    y_test_pre = clf.predict(X_test_rm)
    y_outlier_pre = clf.predict(X_outlier_rm)
    
    cm = ConfusionMatrix(Y_test, y_test_pre, Y_outlier_re, y_outlier_pre)
    index = evaluate(cm)
    Rank[features.keys()[f]] = index
```

```python
Rank =sorted(Rank.items(), key=lambda x: x[1], reverse=False)
```

## Plot confusion matrix

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
plt.figure()
target_names = ['passing', 'failing']
confusion_mat_re = np.zeros((2,2), dtype=int)
confusion_mat_re[0][0] = int(confusion_mat[1][1])
confusion_mat_re[0][1] = int(confusion_mat[1][0])
confusion_mat_re[1][0] = int(confusion_mat[0][1])
confusion_mat_re[1][1] = int(confusion_mat[0][0])
plot_confusion_matrix(confusion_mat_re, classes=target_names,normalize=False,
                    title="confusion matrix")

plt.show()
```
